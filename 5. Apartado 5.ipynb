```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
misdatos = load_breast_cancer()
```

```python
# Establishing our attributes and classes into X and Y 
X = misdatos.data
Y = misdatos.target
```

```python
# Converting X and Y arrays into Dataframes to make it easy to work with
dfx = pd.DataFrame(X)
dfy = pd.DataFrame(Y)
```

```python
plt.plot(dfy)
plt.show()
```

```python
dfy.plot.hist()
plt.xlabel('0: Beningn - 1: Malignant')
plt.ylabel('# Samples')
plt.show()
```
